```python
import requests
import pandas as pd
import schedule
import time
import threading

url = "https://opendata.emel.pt/offstreetparking/lots/availability?limit=1000000"
headers = {
    "api-key": "xxxxxxxxx"
}

def fetch_and_save_data():
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        df = pd.json_normalize(data)
        df.to_csv('parking_data.csv', mode='a', header=False, index=False)
        print("Dados salvos com sucesso!")
    else:
        print(f"Erro: {response.status_code}")

def job():
    threading.Thread(target=fetch_and_save_data).start()


schedule.every(2).minutes.do(job)

while True:
    schedule.run_pending()
    time.sleep(1)
```
